```python
import bs4 as bs
from openpyxl import load_workbook
import pandas as pd
import urllib.request
xls = pd.ExcelFile(r'C:\Users\satwi\Desktop\GRA\P4.xlsx')
df = pd.read_excel(xls, sheet_name="Old Sample-Fac")
urls = df['Final URL'] # Name of the URL column
workbook = load_workbook(filename=r'C:\Users\satwi\Desktop\GRA\P4.xlsx')
sheet = workbook['Old Sample-Fac'] # Name of the required worksheet
i=2
# Here 20,21,22 are column numbers of the values to be inserted
for url in urls:
    try:
        response = urllib.request.urlopen(url)
        html = response.read()
        soup = bs.BeautifulSoup(html)
        if len(soup.find_all(class_='field-label')) != 0:
            for tags in soup.find_all(class_='field-label'):
                # TODO: case insensitive search
                if tags.text.find('INCIDENTS/INJURIES:') != -1 or tags.text.find('Incidents/Injuries:') != -1 or tags.text.find('Accidents/Injuries:') != -1 or tags.text.find('Incidents/Injury:') != -1:
                    print("div", url,i)
                    sheet.cell(row=i, column=34).value = tags.next_sibling.next_sibling.text
                if tags.text.find('Manufacturer(s):') != -1 or tags.text.find('Manufacturer/Importer:') != -1  or tags.text.find('Manufacturer:') != -1:
                    print("div", url,i)
                    sheet.cell(row=i, column=35).value = tags.find_next_sibling("div").text
                if tags.text.find('Manufacturer/Distributor:') != -1 or tags.text.find('Manufacturers:') != -1:
                    print("div", url, i)
                    sheet.cell(row=i, column=35).value = tags.find_next_sibling("div").text
                if tags.text.find('Manufactured in:') != -1 :
                    print("div", url, i)
                    sheet.cell(row=i, column=36).value =  tags.find_next_sibling("div").text
                if tags.text.find('Manufactured In:') != -1 or tags.text.find('Assembled in:') != -1:
                    print("div", url, i)
                    sheet.cell(row=i, column=36).value =  tags.find_next_sibling("div").text
                if tags.text.find('Manufactured:') != -1 :
                    print("div", url, i)
                    sheet.cell(row=i, column=36).value =  tags.find_next_sibling("div").text
        elif len(soup.find_all('strong')) != 0:
            for tags in soup.find_all('strong'):
                if tags.text.find('INCIDENTS/INJURIES:') != -1 or tags.text.find('Incidents/Injuries:') != -1 or tags.text.find('Accidents/Injuries:')!= -1:
                    print("strong", url,i)
                    sheet.cell(row=i, column=34).value = tags.next_sibling
                if tags.text.find('Manufacturer:') != -1 or tags.text.find('Manufacturer(s):') != -1 or tags.text.find('Manufacturer/Importer:') != -1:
                    print("strong", url, i)
                    sheet.cell(row=i, column=35).value = tags.next_sibling
                if tags.text.find('Manufacturer/Distributor:') != -1:
                    print("strong", url, i)
                    sheet.cell(row=i, column=35).value = tags.next_sibling
                if tags.text.find('Manufactured in:') != -1 :
                    print("strong", url,i)
                    sheet.cell(row=i, column=36).value = tags.next_sibling
                if tags.text.find('Manufactured In:') != -1 :
                    print("strong", url,i)
                    sheet.cell(row=i, column=36).value = tags.next_sibling
                if tags.text.find('Manufactured:') != -1 :
                    print("strong", url,i)
                    sheet.cell(row=i, column=36).value = tags.next_sibling
    except:
        # Handle page not found exception
        pass
    if sheet.cell(row=i, column=34).value is None:
        sheet.cell(row=i, column=34).value = "No incident/injuries. Please re-check!"
    if sheet.cell(row=i, column=35).value is None:
        sheet.cell(row=i, column=35).value = "No manufacturers. Please re-check!"
    if sheet.cell(row=i, column=36).value is None:
        sheet.cell(row=i, column=36).value = "No manufacturer country. Please re-check!"
    workbook.save(filename=r'C:\Users\satwi\Desktop\GRA\P4.xlsx')
    i += 1
   # print(url)
```
